```python
from sklearn.metrics import r2_score
from scipy.optimize import curve_fit
import numpy as np
import pandas as pd

# Define the functions
def quadratic(x, a, b, c):
    return a * x**2 + b * x + c

def exponential(x, a, b, c):
    return a * np.exp(b * x) + c

def inverse(x, a):
    return 1 / (a * x)

def sinusoidal(x, a, b):
    return a * np.sin(x) + b

def tangential(x, a, b):
    return a * np.tan(x) + b

def absolute_value(x, a, b, c):
    return a * np.abs(x) + b * x + c

def cociente_polinomios(x, a, b, c):
    return (a * x**2 + b) / (c * x)

def logarithmic(x, a, b):
    return a * np.log(x) + b

def linear_product(x, a, b, c):
    return a * x + b * x + c * x

def quadratic_inverse(x, a):
    return 1 / (a * x**2)

def polynomial_inverse(x, a, b, c):
    return a / (b * x**2) + c * x

# Dictionary of functions
functions = {
    "quadratic": quadratic,
    "exponential": exponential,
    "inverse": inverse,
    "sinusoidal": sinusoidal,
    "tangential": tangential,
    "absolute_value": absolute_value,
    "cociente_polinomios": cociente_polinomios,
    "logarithmic": logarithmic,
    "linear_product": linear_product,
    "quadratic_inverse": quadratic_inverse,
    "polynomial_inverse": polynomial_inverse,
}

# Dictionary of independent variables
independent_vars = {
    "host_response_rate": df['host_response_rate'],
    "host_acceptance_rate": df['host_acceptance_rate'],
    "host_total_listings_count": df['host_total_listings_count'],
    "accommodates": df['accommodates'],
    "reviews_per_month": df['reviews_per_month'],
}

# Loop through each function and independent variable
for func_name, func in functions.items():
    for var_name, var in independent_vars.items():
        # Convert independent variable to numeric
        var = pd.to_numeric(var, errors='coerce')

        # Fit the model
        try:
            parametros, _ = curve_fit(func, var, df['price'])
        except (RuntimeError, ValueError) as e:
            print(f"Error fitting {func_name} with {var_name}: {e}")
            continue

        # Calculate fitted values
        yfit = func(var, *parametros)

        # Check for inf or NaN values in yfit
        if np.any(np.isinf(yfit)) or np.any(np.isnan(yfit)):
            print(f"Warning: Fitted values for {func_name} with {var_name} contain inf or NaN values.")
            continue  # Skip to the next iteration

        # Calculate R-squared
        R2 = r2_score(df['price'], yfit)

        # Print results
        print(f"Function: {func_name}, Variable: {var_name}, R-squared: {R2}")

# Fit model for 'reviews_per_month' as dependent variable
for func_name, func in functions.items():
    # Convert independent variable to numeric
    accommodates = pd.to_numeric(df['accommodates'], errors='coerce')

    # Fit the model
    try:
        parametros, _ = curve_fit(func, accommodates, df['reviews_per_month'])
    except (RuntimeError, ValueError) as e:
        print(f"Error fitting {func_name} for 'reviews_per_month': {e}")
        continue

    # Calculate fitted values
    yfit = func(accommodates, *parametros)

    # Check for inf or NaN values in yfit
    if np.any(np.isinf(yfit)) or np.any(np.isnan(yfit)):
        print(f"Warning: Fitted values for {func_name} contain inf or NaN values.")
        continue  # Skip to the next iteration

    # Calculate R-squared
    R2 = r2_score(df['reviews_per_month'], yfit)

    # Print results
    print(f"Function: {func_name}, Variable: accommodates, R-squared: {R2}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Dictionary of independent variables
independent_vars = {
    "host_response_rate": df['host_response_rate'],
    "host_acceptance_rate": df['host_acceptance_rate'],
    "host_total_listings_count": df['host_total_listings_count'],
    "accommodates": df['accommodates'],
    "reviews_per_month": df['reviews_per_month'],
}

results = []

# Loop through each independent variable
for var_name, var in independent_vars.items():
    # Convert independent variable to numeric
    var = pd.to_numeric(var, errors='coerce')

    # Reshape independent variable to a 2D array
    X = var.values.reshape(-1, 1)

    # Create and fit the linear regression model
    model = LinearRegression()
    model.fit(X, df['price'])

    # Make predictions
    y_pred = model.predict(X)

    # Calculate R-squared
    R2 = r2_score(df['price'], y_pred)

    results.append({'Dependent Variable': 'price', 'Independent Variable': var_name,
                    'Model': 'Linear Regression', 'R-squared': R2})

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Print the results
print(results_df)
```
